3) Hemos construido un diccionario que se puede usar para encontrar el resultado si dos booleanos están conectados con el operador y

6) Las cadenas son inmutables, pero ¿por qué funciona lo siguiente? ¿Claramente cambiamos el valor de la cadena llamada string_1? ¿Qué afirmaciones agregaría para aclarar lo que ha sucedido? usar id

```python
string_1 = "text 1"
print(string_1)
print(id(string_1))
string_1 = "text 2"
print(string_1)
print(id(string_1))
```

**RESPUESTA:**
Lo que sucede es que el primer objeto string_1 guarda la cadena de texto, pero cuando volvermos a guardar otra cadena de texto en el mismo objeto, lo que hacemos es crear otro segundo objeto, diferente al primer objeto con el mismo nombre pero con diferente id, lo que sucede con el primer objeto, es que se termina borrando con su respectivo id.
